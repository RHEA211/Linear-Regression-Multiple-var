```python
#Importing the libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression 
```

```python
#Getting the dataset
dataset = pd.read_csv(r'C:\Users\KIIT\Desktop\final.csv')
dataset
```

```python
dataset.columns[dataset.isna().any()]
```

```python
dataset.LotFrontage = dataset.LotFrontage.fillna(dataset.LotFrontage.mean())
dataset
```

```python
#Mapping Salary data into Numerical Values Values


dataset['MSZoning'] = dataset['MSZoning'].map({'RL': 1, 'RM':2, 'FV':3, 'RH':4, 'C (all)':5}).astype(int)

dataset['HouseStyle'] = dataset['HouseStyle'].map({'1Story': 1, '2Story':2, '1.5Fin':3, '1.5Unf':4, '2.5Fin':5, '2.5Unf':6, 'SLvl':7, 'SFoyer':8}).astype(int)

print(dataset.head)
```

```python

X = dataset.drop('SalePrice',axis='columns')
X
```

```python
Y=dataset.SalePrice
Y
```

```python
#Training
model = LinearRegression()
model.fit(X,Y)
```

```python
#Prediction
a=[[20,1,80,9600,1,6,8]]
PredictedmodelResult = model.predict(a)
print(PredictedmodelResult)
#Example: 20	RL	80	9600	1Story	6	8
```
